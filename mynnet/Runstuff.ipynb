```python
import numpy as np
x = 100*np.random.rand(2,20)
print 'x: \n',x
xnorm = x-np.mean(x,1)[:,np.newaxis]
print 'xnorm: \n',xnorm
```

```python
print 'Correlation: '
corr = np.corrcoef(xnorm)
print corr
[w,v] = np.linalg.eig(corr)
print 'Eigs: \n', w
```

```python
cov = np.cov(xnorm)
[w,v] = np.linalg.eig(cov)
print 'Covariance: \n', cov
print 'x*x.T\n',xnorm.dot(xnorm.T)/19
print 'Eigs: \n', w
```

```python
x = 100*np.random.rand(3,5)
print 'x: \n',x
xnorm = x-np.mean(x,0)[np.newaxis,:]
print 'xnorm: \n',xnorm
```

```python
Cov = np.dot(x.T,x)
[w,v] = np.linalg.eig(Cov)
w = np.maximum(1e-8,w)
x_rot = x.dot(v.dot(np.diag(w**-.5)))
print x_rot.T.dot(x_rot)
```
